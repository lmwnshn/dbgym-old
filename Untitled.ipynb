```python
node_types = [
    "Aggregate",
    "Append",
    "Bitmap Heap Scan",
    "Bitmap Index Scan",
    "Function Scan",
    "Hash",
    "Hash Join",
    "Index Scan",
    "Index Only Scan",
    "Limit",
    "LockRows",
    "Memoize",
    "Merge Join",
    "ModifyTable",
    "Nested Loop",
    "ProjectSet",
    "Result",
    "Seq Scan",
    "Subquery Scan",
    "Sort",
    "WindowAgg",
]

def featurize(
    Node_Type,
    Plan_Width,
    Plan_Rows,
    Total_Cost,
    Join_Type,
    Parent_Relationship,
    Hash_Buckets,
    Sort_Key,
    Sort_Method,
    Relation_Name,
    Attribute_Mins,
    Attribute_Medians,
    Attribute_Maxs,
    Index_Name,
    Scan_Direction,
    Strategy,
    Partial_Mode
):
    node_type = node_types[Node_Type.index(1)]
    features = []
    # Refer to Table 2: QPP Net Inputs.
    
    # All.
    features.append(Plan_Width)
    features.append(Plan_Rows)
    # TODO(WAN): plan buffers, estimated I/Os.
    features.append(Total_Cost)
    
    # Joins.
    if node_type in ["Hash Join", "Merge Join"]:
        features.append(Join_Type)
        features.append(Parent_Relationship)
    # Hash.
    if node_type in ["Hash"]:
        features.append(Hash_Buckets)
        # TODO(WAN): Hash Algorithm
    # Sort.
    elif node_type in ["Sort"]:
        features.extend(Sort_Key)
        features.extend(Sort_Method)
    # Scans.
    elif node_type in ["Seq Scan", "Bitmap Heap Scan"]:
        features.extend(Relation_Name)
        features.extend(Attribute_Mins)
        features.extend(Attribute_Medians)
        features.extend(Attribute_Maxs)
    # Index scans.
    elif node_type in ["Index Scan", "Index Only Scan"]:
        features.extend(Relation_Name)
        features.extend(Attribute_Mins)
        features.extend(Attribute_Medians)
        features.extend(Attribute_Maxs)
        features.extend(Index_Name)
        features.extend(Scan_Direction)
    elif node_type in ["Bitmap Index Scan"]:
        features.extend(Index_Name)
    # Aggregates.
    elif node_type in ["Aggregate"]:
        features.extend(Strategy)
        features.extend(Partial_Mode)
        # TODO(WAN): Operator
    return features

start = time.time()
# Manually unrolling apply for about 2x speedup (45s to 22s on 1.2m rows).
features = np.array(list(map(
    featurize,
    df["Node Type"],
    df["Plan Width"],
    df["Plan Rows"],
    df["Total Cost"],
    df["Join Type"],
    df["Parent Relationship"],
    df["Hash Buckets"],
    df["Sort Key"],
    df["Sort Method"],
    df["Relation Name"],
    df["Attribute Mins"],
    df["Attribute Medians"],
    df["Attribute Maxs"],
    df["Index Name"],
    df["Scan Direction"],
    df["Strategy"],
    df["Partial Mode"],
)), dtype=object)
end = time.time()
print(f"{end - start:.2f} elapsed s")
```

```python
start = time.time()
features_indexes = list(map(lambda obs_idx, children_idxs: np.hstack([[obs_idx], children_idxs]).tolist(), df["Observation Index"], df["Children Indexes"]))
end = time.time()
print(f"{end - start:.2f} elapsed s")

start = time.time()
feature_vectors = [[item for sublist in features[input_vec].tolist() for item in sublist] for input_vec in features_indexes]
end = time.time()
print(f"{end - start:.2f} elapsed s")
```

```python
def pad(node_type, features):
    max_len = max_feature_len[node_type]
    return np.pad(features, (0, max_len - len(features)))

start = time.time()
data["Features"] = list(map(pad, data["Node Type"], data["Features"]))
end = time.time()
print(f"{end - start:.2f} elapsed s")
```

```python
import torch
import torch.nn as nn

class NeuralUnit(nn.Module):
    """Define a Resnet block"""
    
    # Neural networks
    # Unless otherwise stated, 
    # each neural unit had 5 hidden layers, each with 128 neurons each.
    # The data output size was set to d = 32.
    # Rectified linear units were used as activation functions.

    NUM_HIDDEN_LAYERS = 5
    NUM_NEURONS = 128
    DATA_OUTPUT_SIZE = 32

    def __init__(
        self,
        node_type,
        dim_dict,
        num_hidden_layers=NUM_HIDDEN_LAYERS,
        hidden_size=NUM_NEURONS,
        output_size=DATA_OUTPUT_SIZE,
        norm_enabled=False,
    ):
        super().__init__()
        self.node_type = node_type
        self.dense_block = self.build_block(
            num_hidden_layers, hidden_size, output_size, input_dim=dim_dict[node_type]
        )

    def build_block(self, num_layers, hidden_size, output_size, input_dim):
        """Construct a block consisting of linear Dense layers.
        Parameters:
            num_layers  (int)
            hidden_size (int)           -- the number of channels in the conv layer.
            output_size (int)           -- size of the output layer
            input_dim   (int)           -- input size, depends on each node_type
            norm_layer                  -- normalization layer
        Returns a conv block (with a conv layer, a normalization layer, and a non-linearity layer (ReLU))
        """
        
        # 
        
        assert num_layers >= 2
        dense_block = [nn.Linear(input_dim, hidden_size), nn.ReLU()]
        for i in range(num_layers - 2):
            dense_block += [nn.Linear(hidden_size, hidden_size), nn.ReLU()]
        dense_block += [nn.Linear(hidden_size, output_size), nn.ReLU()]

        for layer in dense_block:
            try:
                nn.init.xavier_uniform_(layer.weight)
            except:
                pass
        return nn.Sequential(*dense_block)

    def forward(self, x):
        """Forward function"""
        out = self.dense_block(x)
        return out
```
