```python
from sqlalchemy import create_engine, inspect
import pandas as pd
import numpy as np
from dbgym.util.sql import substitute

import matplotlib.pyplot as plt
from models.qppnet import QPPNet
from math import ceil
from pathlib import Path
import re
```

```python
def get_paths(experiment, iteration, artifact_folder=Path("./artifact/")):
    train = artifact_folder / f"gym/workload.train.db"
    if experiment == "Test":
        test = artifact_folder / f"gym/workload.test.db"
    else:
        test = artifact_folder / f"experiment/transformed_workloads/workload_{experiment}.db"
    obs = artifact_folder / f"qppnet/observations/{experiment}_iteration_{iteration}.parquet"
    return train, test, obs

def get_observations(experiment, iteration, artifact_folder=Path("./artifact/")):
    train, test, obs = get_paths(experiment, iteration, artifact_folder)
    obs_pq = pd.read_parquet(obs)

    queries = []
    for db in [train, test]:
        engine = create_engine(f"sqlite:///{db}")
        df = pd.read_sql(f"SELECT w.query_num, qt.template, w.template_id, w.params_id FROM workload w, query_template qt WHERE w.template_id = qt.id", engine)
        df["query_num"] = df["query_num"].astype(int)
        df["template"] = df["template"].astype("category")
        df["template_id"] = df["template_id"].astype(int)
        df["params_id"] = df["params_id"].astype("Int64")
        queries.append(df)
    queries = pd.concat(queries).set_index("query_num")

    df = obs_pq.join(queries, on="Query Num")
    return df

df = get_observations("Test", 0)
```

```python
name_regex = re.compile(r"^id_(?P<operator>.+)-(?P<num_children>\d+)_epoch_(?P<epoch_num>\d+)_m_validationloss_(?P<validation_loss>\d+.*\d+)_end.pth$")

def get_models(artifact_folder=Path("./artifact/")):
    dfs = {}
    experiments = (artifact_folder / "/qppnet/model/").glob("*")
    for model_folder in experiments:
        rows = []
        for model in model_folder.glob("*.pth"):
            match = name_regex.match(model.name)
            assert match is not None
            match = match.groupdict()
            rows.append(match)
        df = pd.DataFrame(rows)
        df["num_children"] = df["num_children"].astype(int)
        df["epoch_num"] = df["epoch_num"].astype(int)
        df["validation_loss"] = df["validation_loss"].astype(float)
        dfs[model_folder] = df
    return dfs

def evaluate_models(models, test_df):
    eval_dfs = {}
    for model_folder, df in models.items():
        df_tmp = df[["epoch_num", "validation_loss"]].drop_duplicates().sort_values("epoch_num")
        best_idx = df_tmp["validation_loss"].idxmin()
        best_row = df.iloc[best_idx]
        best_epoch = best_row["epoch_num"]

        qppnet = QPPNet(train_df=None, test_df=test_df, save_folder=model_folder)
        qppnet.load_weights(best_epoch)

        eval_dfs[model_folder] = qppnet.evaluate()
        print(model_folder, best_epoch)
        print(qppnet.compute_metrics(eval_dfs[model_folder]))
    return eval_dfs
        
test_df = pd.read_parquet("./artifact/qppnet/observations/Test_iteration_0.parquet")
test_df["Query Hash"] = test_df["Query Hash"].apply(tuple)
```
