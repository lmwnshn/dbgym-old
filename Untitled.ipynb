```python
from sqlalchemy import create_engine, inspect
import pandas as pd
import numpy as np
from dbgym.util.sql import substitute

import matplotlib.pyplot as plt
from models.qppnet import QPPNet
from math import ceil
from pathlib import Path
import re
```

```python
def get_paths(experiment, iteration, artifact_folder=Path("./artifact/")):
    train = artifact_folder / f"gym/workload.train.db"
    if experiment == "Test":
        test = artifact_folder / f"gym/workload.test.db"
    else:
        test = artifact_folder / f"experiment/transformed_workloads/workload_{experiment}.db"
    obs = artifact_folder / f"qppnet/observations/{experiment}_iteration_{iteration}.parquet"
    return train, test, obs

def get_observations(experiment, iteration, artifact_folder=Path("./artifact/")):
    train, test, obs = get_paths(experiment, iteration, artifact_folder)
    obs_pq = pd.read_parquet(obs)

    queries = []
    for db in [train, test]:
        if not db.exists():
            print(f"WARNING: couldn't find {db}")
            continue
        engine = create_engine(f"sqlite:///{db}")
        df = pd.read_sql(f"SELECT w.query_num, qt.template, w.template_id, w.params_id FROM workload w, query_template qt WHERE w.template_id = qt.id", engine)
        df["query_num"] = df["query_num"].astype(int)
        df["template"] = df["template"].astype("category")
        df["template_id"] = df["template_id"].astype(int)
        df["params_id"] = df["params_id"].astype("Int64")
        queries.append(df)
    queries = pd.concat(queries).set_index("query_num")

    df = obs_pq.join(queries, on="Query Num")
    return df
```

```python
name_regex = re.compile(r"^id_(?P<operator>.+)-(?P<num_children>\d+)_epoch_(?P<epoch_num>\d+)_m_validationloss_(?P<validation_loss>\d+.*\d+)_end.pth$")

def get_models(artifact_folder=Path("./artifact/")):
    dfs = {}
    experiments = (artifact_folder / "qppnet/model/").glob("*")
    for model_folder in experiments:
        rows = []
        for model in model_folder.glob("*.pth"):
            match = name_regex.match(model.name)
            assert match is not None
            match = match.groupdict()
            rows.append(match)
        df = pd.DataFrame(rows)
        df["num_children"] = df["num_children"].astype(int)
        df["epoch_num"] = df["epoch_num"].astype(int)
        df["validation_loss"] = df["validation_loss"].astype(float)
        dfs[model_folder] = df
    return dfs

def evaluate_models(models, test_df):
    eval_dfs = {}
    for model_folder, df in models.items():
#         if not any([str(model_folder).endswith(x) for x in ["PerfectPrediction"]]):
#             continue
        df_tmp = df[["epoch_num", "validation_loss"]].drop_duplicates().sort_values("epoch_num")
        best_idx = df_tmp["validation_loss"].idxmin()
        best_row = df.iloc[best_idx]
        best_validation_epoch = best_row["epoch_num"]
        last_epoch = df["epoch_num"].max()
        print(f"Evaluating: {model_folder}")

        eval_dfs[model_folder] = {}
        for epoch_name, epoch in [("best_validation_epoch", best_validation_epoch), ("last_epoch", last_epoch)]:
            save_path = Path("./artifact/evaluate/") / model_folder.name / str(epoch)
            save_path.mkdir(parents=True, exist_ok=True)
            predictions_parquet = save_path / "predictions.parquet"
            
            if predictions_parquet.exists():
                predictions = pd.read_parquet(predictions_parquet)
            else:
                qppnet = QPPNet(train_df=None, test_df=test_df, save_folder=model_folder)
                qppnet.load_weights(epoch)
                predictions = qppnet.evaluate()
            
            metrics = QPPNet.compute_metrics(predictions)
            eval_dfs[model_folder][epoch_name] = {}
            eval_dfs[model_folder][epoch_name]["epoch"] = epoch
            eval_dfs[model_folder][epoch_name]["predictions"] = predictions
            eval_dfs[model_folder][epoch_name]["metrics"] = metrics
            print("\t", epoch_name.ljust(25), *[f"{x:.3f}" for x in metrics])
    return eval_dfs

def augment_evaluations(eval_dfs):
    dfs = {}
    for model_path in eval_dfs:
        model_name = model_path.name
        for epoch_name in eval_dfs[model_path]:
            qppnet_df = eval_dfs[model_path][epoch_name]["predictions"].set_index("Observation Index")
            obs_df = get_observations(model_name, iteration=0)
            df = obs_df.join(qppnet_df, on="Observation Index", how="inner")
            dfs[(model_name, epoch_name)] = df
    return dfs
```

```python
test_df = pd.read_parquet("./artifact/qppnet/observations/Test_iteration_0.parquet")
test_df["Query Hash"] = test_df["Query Hash"].apply(tuple)

models = get_models(artifact_folder=Path("./artifact/"))
eval_dfs = evaluate_models(models, test_df)
```

```python
augmented_dfs = augment_evaluations(eval_dfs)
for model_name, df in augmented_dfs.items():
    pdf = df.copy()
    logplot = True
    pdf["diff (us)"] = (pdf["Estimated Latency (us)"] - pdf["Actual Latency (us)"]).abs()
    ax = pdf.hist(column="Actual Latency (us)", by="Node Type", figsize=(14,8), log=logplot, bins=20, alpha=0.5, color="b")
    plot_ax = ax.ravel()[:df["Node Type"].nunique()]
    ax = pdf.hist(ax=plot_ax, column="Estimated Latency (us)", by="Node Type", figsize=(14,8), log=logplot, bins=20, alpha=0.5, color="r")
    print(model_name)
    plt.show()
```

```python
summary = []
for model_path, data in eval_dfs.items():
    for key in ["best_validation_epoch", "last_epoch"]:
        summary.append((model_path.name, data[key]["epoch"], *data[key]["metrics"]))
summary = pd.DataFrame(summary, columns=["Name", "Epoch", "RMSE", "RE", "MAE", "RQ"])
summary.sort_values("RMSE")
```
